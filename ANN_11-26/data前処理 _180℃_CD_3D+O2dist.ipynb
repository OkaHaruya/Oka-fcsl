```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame(columns=['x座標','y座標','z座標','V[V]','i[A/m^2]','c[mol/m^3]'])
```

```python
def data_add(f1,f2,v):
    global df
    df1 = pd.read_excel(f1,names=['x座標','y座標','z座標','i[A/m^2]'] ,header=8)
    df2 = pd.read_excel(f2,names=['x座標','y座標','z座標','c[mol/m^3]'] ,header=8)
    df3 = df1.merge(df2, on=['x座標','y座標','z座標'], how="outer")
    df3.insert(3, 'V[V]', v)
    df4 = df3.fillna(0)
    df = pd.concat([df, df4])
    return df  
```
