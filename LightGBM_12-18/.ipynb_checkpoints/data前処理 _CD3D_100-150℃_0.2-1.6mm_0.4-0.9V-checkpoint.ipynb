```python
import numpy as np
import pandas as pd
import glob
import re
```

```python
path = "/Users/aruhy/OneDrive/デスクトップ/data/12-12"
```

```python
df = pd.DataFrame(columns=['x座標','y座標','z座標','V[V]','T_cell[℃]','U_cath[m/s]','i[A/m^2]'])
U_range = ["*_1.0ms","*_1.5ms","*_2.0ms"]
for U in U_range:
    u = re.findall(r'_(.*?)ms', U)
    T_range = ["/*_100℃"+U,"/*_110℃"+U,"/*_120℃"+U,"/*_130℃"+U,"/*_140℃"+U,"/*_150℃"+U]
    for T in T_range:
        files1 = glob.glob(path + T)
        for filename in files1:
            df1 = pd.read_csv(filename,names=['x座標','y座標','z座標','i[A/m^2]'], delim_whitespace=True, header=8)
            v = re.findall(r'_3D(.*?)V_',filename)
            df1.insert(3,'V[V]', v[0])
            t = re.findall(r'V_(.*?)℃_',filename)
            df1.insert(4,'T_cell[℃]',t[0])
            df1.insert(5,'U_cath[m/s]',u[0])
            df = pd.concat([df, df1])
df['V[V]'] = df['V[V]'].astype(float)
df['T_cell[℃]'] = df['T_cell[℃]'].astype(float)
df['U_cath[m/s]'] = df['U_cath[m/s]'].astype(float)
df
```
