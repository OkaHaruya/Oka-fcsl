```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame(columns=['x座標','y座標','V[V]','i[A/m^2]'])
```

```python
def data_add(f,v):
    global df
    df1 = pd.read_excel(f,names=['x座標','y座標','i[A/m^2]'] ,header=8)
    df1.insert(2, 'V[V]', v)
    df = pd.concat([df, df1])
    return df  
```
