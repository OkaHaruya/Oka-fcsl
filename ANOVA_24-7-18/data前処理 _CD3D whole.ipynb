```python
import numpy as np
import pandas as pd
import glob
import re
from sklearn.preprocessing import StandardScaler, MinMaxScaler
import tqdm
```

```python
path = "/Users/user/Desktop/Oka/COMSOL/PEFC/data/24-7-19"
#path ="/Users/aruhy/OneDrive/デスクトップ/data/1-25"
```

```python
df = pd.DataFrame(columns=['x座標','y座標','z座標','V[V]','i[A/m^2]'])

#CD_3D 110℃_0.4V
files = "/*"
files2 = glob.glob(path+files)
for filename in tqdm.tqdm(files2):
            
    df1 = pd.read_csv(filename,names=['x座標','y座標','z座標','i[A/m^2]'], delim_whitespace=True, header=8)
    v = re.findall(r'℃_(.*?)V',filename)
    df1.insert(3,'V[V]', v[0])
    df = pd.concat([df, df1])
        
    
df['V[V]'] = df['V[V]'].astype(float)

df.reset_index(drop=True, inplace=True)
df
```
